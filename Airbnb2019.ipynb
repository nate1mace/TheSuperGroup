```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

source = 'AirbnbSept19.csv'
df = pd.read_csv(source)
```

```python
df['price'] = df['price'].astype(float)
df.dtypes
df
```

```python
df_neighborhood = df[['neighbourhood_group','neighbourhood','price']]
#df_neighborhood = df_neighborhood.groupby(['neighbourhood_group','neighbourhood']).mean()

df_borough = df_neighborhood.groupby(['neighbourhood_group']).mean().reset_index()
```

```python
# df_neighborhood.head()
df_borough.head()
```

```python
# plt.bar(df_neighborhood['neighbourhood'],df_neighborhood['price'])
# df_neighborhood=df_neighborhood.astype(float)
# plot = df_neighborhood.plot.bar(x='price')
plt.figure(figsize=(10,4))
plt.bar(df_borough['neighbourhood_group'],df_borough['price'],)
plt.title("Average Airbnb Cost by NYC Borough, 2019", size = (20))
plt.ylabel("Cost, One Night ($)", size = (16))
plt.xlabel('Borough', size = (16))

plt.savefig("Airbnb Cost NYC 2019")
```
