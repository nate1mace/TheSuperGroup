```python
#Focusing only on Brooklyn for Airbnb data as the csv file for NYC Rolling Sales is by borough
df_bk_nbrhd = df_nbrhd.loc['Brooklyn'].reset_index()

#Renaming Airbnb_brooklyn_sept
df_abb_bk_sept = df_bk_nbrhd
df_abb_bk_sept = df_abb_bk_sept.rename(columns = {'neighbourhood':'NEIGHBORHOOD','price':'AirBnB Price'})

df_abb_bk_sept
```

```python
#DF2 is the NYC Rolling Sales, this was pulled from the csv in the first box (source2)
df2[' SALE PRICE '] = df2[' SALE PRICE '].str.replace(',','').astype(float)

#Create a boolean mask to sort by the month of September (Airbnb CSV is just September, we can discuss if this step is necessary together)
mask = (df2['SALE DATE'] > '9/0/2019') & (df2['SALE DATE'] <= '9/31/2019')
df2_sept = df2.loc[mask]
```

```python
#The table had a lot of sales at 0. Not sure why.. removed
df2_nonzero_sept = df2_sept.loc[df2[' SALE PRICE '] > 0]

#Group NYC Sales data by average price by neighborhood
df2_nbrhd_sept = df2_nonzero_sept.groupby("NEIGHBORHOOD").agg({' SALE PRICE ':'mean'}).reset_index()
df2_nbrhd_sept_count = df2_nonzero_sept.groupby("NEIGHBORHOOD").agg({' SALE PRICE ':'count'}).reset_index()

#Make the neighborhoods into title case so they can be joined
df2_nbrhd_sept['NEIGHBORHOOD'] = df2_nbrhd_sept['NEIGHBORHOOD'].str.lower()
df2_nbrhd_sept['NEIGHBORHOOD'] = df2_nbrhd_sept['NEIGHBORHOOD'].str.title()
df2_nbrhd_sept_count
```

```python
#Just wanted to see all the rows, it was giving me a truncated table
pd.set_option("display.max_rows", 101) 

#Merge both tables, on neighborhood. If you scroll down, you can see that there are many neighborhoods that are not matching
df_bk_sept_19 = pd.merge(df_abb_bk_sept,df2_nbrhd_sept, on = 'NEIGHBORHOOD',how = 'outer')
df_bk_sept_19
```
